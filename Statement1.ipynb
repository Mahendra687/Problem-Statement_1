#### Problem Statement_1

We have data of product sales in the 'Dataset' sheet. Below are your tasks:				
1	Draw a Pivot table to get the different types of outlets and their counts			
2	Draw a Pivot table to get the total sales for different outlets							
3	Draw a Pivot table to find out the total sales of different item types in different outlet types											
4	In the third Pivot table, filter out (remove) low fat products 							
5	Add slicers for item fat and item type in the last Pivot table											

```python
from pandas import*
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
sales = read_excel("sales.xlsx")
sales.head()
```

1. Draw a Pivot table to get the different types of outlets and their counts

```python
pivot = sales.pivot_table(index=['Outlet_Type'])
print(pivot)
```

```python
out_type = sales['Outlet_Type'].value_counts()
out_type
```

2. Draw a Pivot table to get the total sales for different outlets

```python
total_sales = pd.pivot_table(data=sales, index =['Outlet_Type'], 
                             values =['Item_Outlet_Sales'], aggfunc ='sum')
total_sales
```

3. Draw a Pivot table to find out the total sales of different item types in different outlet types

```python
item_types = pd.pivot_table(data=sales,index =["Outlet_Type","Item_Type"], values =['Item_Outlet_Sales'], 
                                aggfunc ='sum')
item_types
```

4. In the third Pivot table, filter out (remove) low fat products

```python
table = pd.pivot_table(data=sales,index=['Item_Fat_Content'])
table
```

```python
options = ['Regular'] 
# selecting rows based on condition 
filter = sales[sales['Item_Fat_Content'].isin(options)] 
filter.head()
```

```python
remove = pd.pivot_table(data=filter,index =["Outlet_Type","Item_Type","Item_Fat_Content"], 
                        values =['Item_Outlet_Sales'], aggfunc ='sum')
remove.head()
```
